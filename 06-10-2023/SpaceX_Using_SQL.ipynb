### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
query = """
    SELECT SUM(PayloadMass) AS Masa_Total
    FROM SPACEXTBL
    WHERE Customers = 'NASA (CRS)'
"""

# Ejecuta la consulta y carga el resultado en un DataFrame
df = pd.read_sql(query, con=engine)

# Muestra la masa total transportada por lanzamientos de NASA (CRS)
print("Masa total transportada por lanzamientos de NASA (CRS):")
print(df)
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query = """
    SELECT AVG(PayloadMass) AS Masa_Media
    FROM SPACEXTBL
    WHERE Serial LIKE 'B1%'
"""

# Ejecuta la consulta y carga el resultado en un DataFrame
df = pd.read_sql(query, con=engine)

# Muestra la masa media transportada por lanzamientos con serie que comienza por 'B1'
print("Masa media transportada por lanzamientos con serie que comienza por 'B1':")
print(df)
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query = """
    SELECT MIN(Date) AS Fecha_Primer_Aterrizaje_RTLS
    FROM SPACEXTBL
    WHERE Mission_Outcome = True
    AND Class = 'TRUE'
"""

# Ejecuta la consulta y carga el resultado en un DataFrame
df = pd.read_sql(query, con=engine)

# Muestra la fecha del primer aterrizaje exitoso sobre terreno (TRUE RTLS)
print("Fecha del primer aterrizaje exitoso sobre terreno (TRUE RTLS):")
print(df)
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query = """
    SELECT DISTINCT Serial
    FROM SPACEXTBL
    WHERE LandingOutcome = 'TRUE ASDS'
    AND PayloadMass BETWEEN 4000 AND 6000
"""

# Ejecuta la consulta y carga los resultados en un DataFrame
df = pd.read_sql(query, con=engine)

# Muestra las series de cohetes que cumplen con los criterios
print("Series de cohetes que aterrizaron sobre portaviones (TRUE ASDS) con masa entre 4000 y 6000:")
print(df)




```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
query_exitosos = """
    SELECT COUNT(*) AS Lanzamientos_Exitosos
    FROM SPACEXTBL
    WHERE Class = True
"""

# Consulta SQL para contar el número de lanzamientos fallidos
query_fallidos = """
    SELECT COUNT(*) AS Lanzamientos_Fallidos
    FROM SPACEXTBL
    WHERE Class = False
"""

# Ejecuta las consultas y carga los resultados en DataFrames
df_exitosos = pd.read_sql(query_exitosos, con=engine)
df_fallidos = pd.read_sql(query_fallidos, con=engine)

# Muestra el número de lanzamientos exitosos y fallidos
print("Número de lanzamientos exitosos:", df_exitosos['Lanzamientos_Exitosos'][0])
print("Número de lanzamientos fallidos:", df_fallidos['Lanzamientos_Fallidos'][0])
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python
query_max_mass = """
    SELECT MAX(PayloadMass) AS Masa_Maxima
    FROM SPACEXTBL
"""

# Ejecuta la consulta para encontrar la masa máxima
df_max_mass = pd.read_sql(query_max_mass, con=engine)

# Obtiene el valor de la masa máxima
masa_maxima = df_max_mass['Masa_Maxima'][0]

# Consulta SQL para obtener las series de cohetes que transportaron la masa máxima
query_series_max_mass = """
    SELECT DISTINCT Serial
    FROM SPACEXTBL
    WHERE PayloadMass = {}
""".format(masa_maxima)

# Ejecuta la consulta y carga los resultados en un DataFrame
df_series_max_mass = pd.read_sql(query_series_max_mass, con=engine)

# Muestra las series de cohetes que transportaron la masa máxima
print("Series de cohetes que transportaron la masa máxima:")
print(df_series_max_mass)
```

### Ejercicio 9

##### Muestra los el resultados del aterrizaje, la serie del cohete y las plataforma de lanzamiento de aquellos lanzamientos fallidos o no intentados en el año 2015


```python
query = """
    SELECT LandingOutcome, Serial, LaunchSite
    FROM SPACEXTBL
    WHERE (Class = False OR Class IS NULL)
    AND Date = 2015
"""

# Ejecuta la consulta y carga los resultados en un DataFrame
df = pd.read_sql(query, con=engine)

# Muestra los resultados del aterrizaje, la serie del cohete y las plataformas de lanzamiento
print("Resultados del aterrizaje, serie del cohete y plataformas de lanzamiento en 2015:")
print(df)
```

### Ejercicio 10

##### Muestra los resultados y la frecuencia de estos resultados entre 2010-06-04 y 2017-03-20 ordenados en orden descendiente según la frecuencia


```python
query = """
    SELECT LandingOutcome, COUNT(*) AS CountLaunches
    FROM SPACEXTBL
    WHERE Date >= '2010-06-04' AND Date <= '2017-03-20'
    GROUP BY LandingOutcome
    ORDER BY CountLaunches DESC
"""

# Ejecuta la consulta y carga los resultados en un DataFrame
df = pd.read_sql(query, con=engine)

# Muestra los resultados y su frecuencia en orden descendente

print(df)
```
